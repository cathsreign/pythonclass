```python
# board
from IPython.display import clear_output

def display_board(board):
    
    print(board[1] + " | " + board[2] + " | " + board[3])
    print("---------")
    print(board[4] + " | " + board[5] + " | " + board[6])
    print("---------")
    print(board[7] + " | " + board[8] + " | " + board[9])

# marker assignment
def player_input():
    
    p1marker = ''
    
    while p1marker != 'X' or p1marker != 'O':
        p1marker = input("Would you like to be 'X' or 'O'? ").upper()
        
        if p1marker == 'X':
            return ('X', 'O')
            
        elif p1marker == 'O':
            return ('O', 'X')
        
        else:
            print("Sorry, that is not 'X' or 'O'!")

# placing the marker
def place_marker(board, marker, position):
    board[position] = marker
    
# checking for wins
def win_check(board, mark):

    # top horizontal
    return ((board[1] == mark and board[2] == mark and board[3] == mark) or 
    # middle horizontal
    (board[4] == mark and board[5] == mark and board[6] == mark) or
    # bottom horizontal
    (board[7] == mark and board[8] == mark and board[9] == mark) or
    # left vertical
    (board[1] == mark and board[4] == mark and board[7] == mark) or
    # middle vertical
    (board[2] == mark and board[5] == mark and board[8] == mark) or
    # right vertical
    (board[3] == mark and board[6] == mark and board[9] == mark) or
    # diagonal
    (board[3] == mark and board[5] == mark and board[7] == mark) or 
    (board[1] == mark and board[5] == mark and board[9] == mark))

# who will go first
import random

def choose_first():

    if random.randint(1,2) == 1:
        return 'Player 1'

    else:
        return 'Player 2'

# checking free space
def space_check(board, position):
    
    return board[position] == ' '

# checking if board is full
def full_board_check(board):
    
    for marker in range(1,10):
        
        if space_check(board, marker):
            return False
        
    return True

# asking for next position
def player_choice(board):
    
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input('Choose your position (1-9): '))

    return int(position)

# play again
def replay():
    
    choice = input("Would you like to play again? Y or N ")
    
    return choice.upper() == 'Y'
```

```python
print('Welcome to Tic Tac Toe!')

while True:

    # setting up the game
    tttboard = [' '] * 10
    turn = choose_first()
    p1marker, p2marker = player_input()
    print(turn + ' goes first!')
    
    play_game = input('Are you ready to play? Y or N ')
    
    if play_game.lower() == 'y':
        game_on = True
    
    else:
        game_on = False

    while game_on:
        
        if turn == 'Player 1':
            
            # p1 turn
            display_board(tttboard)
            position = player_choice(tttboard)
            place_marker(tttboard, p1marker, position)

            if win_check(tttboard, p1marker):
                display_board(tttboard)
                print('Player 1 wins!')
                game_on = False
                
            else:
                
                if full_board_check(tttboard):
                    display_board(tttboard)
                    print("It's a tie!")
                    break
                    
                else:
                    turn = 'Player 2'

        else:

            # p2 turn
            display_board(tttboard)
            position = player_choice(tttboard)
            place_marker(tttboard, p2marker, position)

            if win_check(tttboard, p2marker):
                display_board(tttboard)
                print('Player 2 wins!')
                game_on = False
            
            else:
                if full_board_check(tttboard):
                    display_board(tttboard)
                    print("It's a tie!")
                    break
        
                else:
                    turn = 'Player 1'

    if replay() == False:
        break
```
